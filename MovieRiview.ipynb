```python
import pandas as pd
```

```python
df=pd.read_csv('IMDB Dataset.csv',sep=',',engine='python',quotechar='"',error_bad_lines=False)
df
```

```python
df.shape
```

```python
import re
REPLACE_NO_SPACE = re.compile("[.;:!\'?,\"()\[\]]")
REPLACE_WITH_SPACE = re.compile("(<br\s*/><br\s*/>)|(\-)|(\/)")
df['review_new'] = df['review'].apply(lambda x : REPLACE_NO_SPACE.sub("", x.lower()))
df['review_new'] = df['review_new'].apply(lambda x : REPLACE_WITH_SPACE.sub(" ", x))
df = df.drop(columns = 'review')
df['review']=df['review_new']
df = df.drop(columns = 'review_new')
df
```

```python
import nltk
from nltk.corpus import stopwords
stop=set(stopwords.words('english'))
stop
```

```python
#removing stopwords from the dataset
df_s=df.copy()
#df_s.columns = ['review','sentiment']
df_s['review']=df_s['review'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
df_s
```
